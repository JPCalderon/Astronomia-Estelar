```python
# Una vez identificados los espectros con lo que se va a trabajar.
# se pasa a graficar cada espectro y caracterizar cada linea
# caracteristica de cada espectro.
#
# En la esquina inferior derecha, aparecera las coordenadas del mouse.

import mpld3
from mpld3 import plugins
import pandas as pd

# Grafica el espectro04.fits
id = 4
fig, ax = plt.subplots()

plt.rcParams['figure.figsize'] = [15, 8]
plt.rcParams.update({'font.size': 20})

plt.title('FITS file with 1D:' + str(target[id]) + '.fits')
plt.xlabel('Longitud de onda ($\AA$)')
plt.ylabel('Flujo arbitrario (erg cm${-2}$ s$^{-1}$ $\AA^{-1}$)')

plt.plot(spec1d[id].spectral_axis, spec1d[id].flux, c='green')
plugins.connect(fig, plugins.MousePosition(fontsize=18, fmt='5.1f'))

mpld3.display()
```
