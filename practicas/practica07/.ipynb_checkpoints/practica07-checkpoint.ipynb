```python
# 3. Visualizar los espectros. Recuerde modificar la variable id.

import matplotlib.pyplot as plt

# Grafica el 4.fits
id = 1 # De 1 a 10

plt.rcParams['figure.figsize'] = [15, 8]
plt.rcParams.update({'font.size': 20})

plt.title('FITS file with 1D: ' + str(target[id]) + '.fits')
plt.xlabel('Longitud de onda ($\AA$)')
plt.ylabel('Flujo arbitrario (erg cm${-2}$ s$^{-1}$ $\AA^{-1}$)')

plt.grid(True, which='both', lw = 2, ls = '--', c = '.75')
plt.plot(spec1d[id].spectral_axis, spec1d[id].flux, c='violet')

plt.show()
```

```python
# 4. Realizar el ajuste del prefil Gaussiano para cada línea espectral.
# En este caso sera necesario modificar id y cen.

from lmfit.models import GaussianModel
from practica07 import gaussian
from practica07 import load_spectra
from numpy import exp 
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from lmfit import Model

id = 1 # De 1 a 10
cen=5895.92 # Modificar el valor de centrado
#Valores en reposo
## He i & 5875.62 \AA
## Na i & 5889.95 \AA 
## Na i & 5895.92 \AA 

# Ajuste de la función Gaussiana, para cada línea, modificar el valor inicial
# de centrado del ajuste (cen). 
gmodel = Model(gaussian)
result = gmodel.fit(spec1d[id].flux.value, x=spec1d[id].spectral_axis.value, amp=-0.8, cen=cen, wid=1)

plt.rcParams['figure.figsize'] = [15, 8]
plt.rcParams.update({'font.size': 20})

plt.title('FITS file with 1D: ' + str(target[id]) + '.fits')
plt.xlabel('Longitud de onda ($\AA$)')
plt.ylabel('Flujo arbitrario (erg cm${-2}$ s$^{-1}$ $\AA^{-1}$)')

plt.plot(spec1d[id].spectral_axis.value, spec1d[id].flux.value, c='blue')
plt.plot(spec1d[id].spectral_axis.value, result.best_fit, c='red')

plt.grid(True, which='both', lw = 2, ls = '--', c = '.75')
plt.show()

# Imprime el resultado del ajuste
print(result.fit_report())
```
